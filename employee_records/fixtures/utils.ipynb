```python
import pandas as pd
```

```python
df=pd.read_csv('employee.csv')
```

```python
df.head()
```

```python
df['first_name'] = df['staff_name'].apply(lambda x: x.split()[0] if isinstance(x, str) else None)
```

```python
df.head()
```

```python
df['last_name'] = df['staff_name'].apply(lambda x: x.split()[1] if isinstance(x, str) else None)
```

```python
df.head()
```

```python
# Change 'team leader' to 'team_leader' and convert job_title to lowercase
df['job_title'] = df['job_title'].str.replace('team leader', 'team_leader').str.lower()

# Remove currency symbol from hourly_rate and convert to lowercase
df['hourly_rate'] = df['hourly_rate'].str.replace('£', '').str.lower()

# Remove first_name and last_name columns
df.drop(['first_name', 'last_name'], axis=1, inplace=True)

# Convert the entire dataset to lowercase
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```
